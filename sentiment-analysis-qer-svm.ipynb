```python
copy_reviews_df = reviews_df.copy(deep=True)
```

```python
copy_reviews_df['ls'] = copy_reviews_df['stopword'].apply(lambda x: x.split(" "))
copy_reviews_df.head()
```

```python
# positive reviews
positive_reviews = copy_reviews_df[copy_reviews_df['sentiment']=='positive']['ls']
print("Total positive reviews: ", len(positive_reviews))
df_positive_document = pd.DataFrame(positive_reviews)
positive_reviews_words = list(itertools.chain(*positive_reviews))
print("Total words in positive reviews:", len(positive_reviews_words))

# negative reviews
negative_reviews = copy_reviews_df[copy_reviews_df['sentiment']=='negative']['ls']
print("Total negative reviews: ", len(negative_reviews))
df_negative_document = pd.DataFrame(negative_reviews)
negative_reviews_words = list(itertools.chain(*negative_reviews))
print("Total words in negative reviews:", len(negative_reviews_words))
```

```python
# convert list formated string to list
import ast

def convert_text_list(texts):
    texts = ast.literal_eval(str(texts))
    return [text for text in texts]

df_positive_document["comment_list"] = df_positive_document['ls'].apply(convert_text_list)
df_negative_document["comment_list"] = df_negative_document['ls'].apply(convert_text_list)
```

```python
def calc_TF(document):
    # Counts the number of times the word appears in review
    TF_dict = {}
    for term in document:
        if term in TF_dict:
            TF_dict[term] += 1
        else:
            TF_dict[term] = 1
    # Computes tf for each word
    for term in TF_dict:
        TF_dict[term] = TF_dict[term] / len(document)
    return TF_dict

df_positive_document["TF_dict"] = df_positive_document['comment_list'].apply(calc_TF)
df_negative_document["TF_dict"] = df_negative_document['comment_list'].apply(calc_TF)

# df_positive_document["TF_dict"].head()
# df_negative_document["TF_dict"].head()
```

```python
def calc_DF(tfDict):
    count_DF = {}
    # Run through each document's tf dictionary and increment countDict's (term, doc) pair
    for document in tfDict:
        for term in document:
            if term in count_DF:
                count_DF[term] += 1
            else:
                count_DF[term] = 1
    return count_DF

DF_positive = calc_DF(df_positive_document["TF_dict"])
DF_negative = calc_DF(df_negative_document["TF_dict"])

print(DF_positive)
print(len(DF_positive))
print(len(DF_negative))
```

```python
# words frequency

positive_words_frequency = collections.Counter(DF_positive)
negative_words_frequency = collections.Counter(DF_negative)

positive_freq_words_df = pd.DataFrame.from_records(list(dict(positive_words_frequency).items()), columns=['word','frequency'])
negative_freq_words_df = pd.DataFrame.from_records(list(dict(negative_words_frequency).items()), columns=['word','frequency'])
positive_freq_words_df
```

```python
df_count_words = pd.concat([positive_freq_words_df, negative_freq_words_df], axis = 1)

df_count_words.drop(['word'], axis=1, inplace=True)
df_count_words.columns = ['positive', 'negative']
df_count_words['positive'] = df_count_words['positive'].fillna(0)
df_count_words['negative'] = df_count_words['negative'].fillna(0)
df_count_words['positive'] = pd.to_numeric(df_count_words['positive'], downcast='integer')
df_count_words['negative'] = pd.to_numeric(df_count_words['negative'], downcast='integer')
df_count_words
```

```python
num_features = len(df_count_words)
num_features
```

```python
df_count_positif = reviews_df[reviews_df["sentiment"] == "positive"]
df_count_negatif = reviews_df[reviews_df["sentiment"] == "negative"]
print("positive: ", len(df_count_positif))
print("negative: ", len(df_count_negatif))
```

```python
n_positif = len(df_count_positif)
n_negatif = len(df_count_negatif)

# mendefinisikan fungsi unyuk menghitung pf, qf, dan qer score
def calculate_pf(df_positif, n_positif):
    pf = (df_positif + 0.5) / (n_positif + 1)
    return pf

def calculate_qf(df_negatif, n_negatif):
    qf = (df_negatif + 0.5) / (n_negatif + 0.5)
    return qf

def score_qer(pf, qf):
    score = np.abs(pf+qf) / np.abs(pf-qf)
    return score

# melakukan perhitungan score qer untuk masing masing feature

score = np.zeros((num_features, 2))


for i in range(num_features):
    df_positif = df_count_words.iloc[i, 0]
    pf = calculate_pf(df_positif, n_positif)
    df_negatif = df_count_words.iloc[i, 1]
    qf = calculate_qf(df_negatif, n_negatif)
    score_qr = score_qer(pf, qf)

    score[i, 0] = i
    score[i, 1] = score_qr
    
# menyimpan hasil perhitungan qer score ke dalam dataframe
df_qer = pd.DataFrame(score, columns=["index", "qer_score"])
df_qer = df_qer.astype({'index':'int'})
df_qer
```

```python
# melihat isi dataframe df_qer yang sudah diurutkan nilai qer dari terbesar ke terendah
print(df_qer.sort_values(by='qer_score', ascending=False))
```

```python
# ratio feature selection

def ratio_feature(percent, num_features):
    result = int((percent/100) * num_features)
    return result

twenty_five_percent = ratio_feature(25, num_features)
fifty_percent = ratio_feature(50, num_features)
seventy_five_percent = ratio_feature(75, num_features)
one_hundred_percent = ratio_feature(100, num_features)
print(one_hundred_percent)
```

```python
# fungsi untuk menghitung akurasi, metrics, report qer

def qer_accuracy(df_qer, num_feature_used, df_tfidf, y):
    df_sorted_qer = df_qer.sort_values(by='qer_score', ascending=False).head(num_feature_used)
    idx_word = df_sorted_qer['index'].values
    X_fix = df_tfidf.iloc[:, list(idx_word)]
    
    X_train, X_test, y_train, y_test = train_test_split(X_fix, y, test_size=0.2, random_state=1)

    linear_svm=SVC(kernel='linear') 
    linear_svm.fit(X_train, y_train)
    predicted = linear_svm.predict(X_test)
    
#     for i in range(len(X_test)):
#         res=linear_svc.predict(X_test)[i]
#         if res=="positive":
#             print("positif")
#         elif res=="negative":
#             print("negatif")
    
#     y_pred_train = linear_svc.predict(X_train)
#     print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
    print('Akurasi = %s ' %accuracy_score(y_test, predicted))
#     print('Presisi = %s ' %precision_score(y_test, predicted, average="binary", pos_label='negative'))
#     print('Recall = %s ' %recall_score(y_test, predicted, average="binary", pos_label='negative'))
#     print('F-measure = %s ' %f1_score(y_test, predicted, average="binary", pos_label='negative'))
```
